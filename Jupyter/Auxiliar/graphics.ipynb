```python
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf

# Carregue o arquivo de áudio (substitua 'seu_audio.wav' pelo caminho do seu arquivo de áudio)
audio, samplerate = sf.read('dataset-16kHz-16b-1s/bom/301.wav')

# Calcule o número de pontos de dados correspondentes a 30ms
num_pontos = int(0.005 * samplerate)

# Crie um vetor de tempo com base no samplerate e no número de pontos
tempo = np.linspace(0, num_pontos / samplerate, num_pontos)

# Extraia a porção de áudio correspondente a 30ms
audio_data = audio[:num_pontos]

# Plote o sinal de áudio
plt.plot(tempo, audio_data, marker='.', linestyle='None', lw=0.5)

# Adicione uma linha vertical para cada ponto discreto
for t, audio_value in zip(tempo, audio_data):
    plt.vlines(t, 0, audio_value, colors='b', linestyles='solid', lw=0.5)  # Ajuste a cor, estilo da linha e largura conforme desejado

plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal de Áudio de 5ms')
plt.grid(True)

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf

# Carregue o arquivo de áudio (substitua 'seu_audio.wav' pelo caminho do seu arquivo de áudio)
audio, samplerate = sf.read('dataset-16kHz-16b-1s/bom/301.wav')

# Calcule a Transformada de Fourier do áudio
espectro = np.fft.fft(audio)
frequencias = np.fft.fftfreq(len(espectro), 1.0 / samplerate)

# Descarte metade do espectro (componentes espelhadas)
espectro = espectro[:len(espectro) // 2]
frequencias = frequencias[:len(frequencias) // 2]

# Plote o espectro de frequência
plt.figure(figsize=(10, 6))  # Ajuste o tamanho do gráfico conforme desejado
plt.plot(frequencias, np.abs(espectro))  # Use np.abs() para obter a magnitude do espectro

# Configurações do eixo
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude do Espectro')
plt.title('Espectro de áudio - Retificador bom')

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import soundfile as sf

# Carregue o arquivo de áudio (substitua 'seu_audio.wav' pelo caminho do seu arquivo de áudio)
audio, samplerate = sf.read('dataset-16kHz-16b-1s/bom/301.wav')

# Calcule o espectrograma usando a função 'spectrogram' do módulo 'signal' do scipy
f, t, Sxx = signal.spectrogram(audio, fs=samplerate)

# Plote o espectrograma
plt.figure(figsize=(10, 6))  # Ajuste o tamanho do gráfico conforme desejado
plt.pcolormesh(t, f, 10 * np.log10(Sxx), shading='auto')  # Use 'shading' para uma melhor representação de cores

# Configurações do eixo
plt.colorbar(label='Potência Espectral (dB)')
plt.xlabel('Tempo (s)')
plt.ylabel('Frequência (Hz)')
plt.title('Espectrograma de Áudio')

# Exiba o gráfico
plt.show()
```
